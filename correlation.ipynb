```python
import numpy as np
import random
np.random.seed(1)
```

```python
x= np.random.randint(0, 50, 1000) #1000 random numbers bw 0 & 50
```

```python
x
```

```python
y= x + np.random.normal(0, 10, 1000)
#positive correlation w some noise
```

```python
y
```

```python
np.corrcoef(x, y)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')

plt.scatter(x, y)
plt.show()
```

```python
y= x - np.random.normal(0, 10, 1000)
#negative correlation w some noise
```

```python
y
```

```python
np.corrcoef(x, y)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')

plt.scatter(x, y)
plt.show()
```

```python
import pandas as pd
data= pd.DataFrame({'a': np.random.randint(0, 50, 1000)})
data['b']= data['a'] + np.random.normal(0, 10, 1000)
data['c']= 100 - data['a'] + np.random.normal(0, 5, 1000)
data['d']= np.random.randint(0, 50, 1000)

data.corr()
```

```python
pd.plotting.scatter_matrix(data, figsize=(6,6))
plt.show()
```

```python
plt.matshow(data.corr())
plt.xticks(range(len(data.columns)), data.columns)
plt.yticks(range(len(data.columns)), data.columns)
plt.colorbar()
plt.show()
```
